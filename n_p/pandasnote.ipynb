 # Series

## 基本

```python
s=pd.Series([1,2,3,4,5,6])
print(s)
#第一列索引，第二列数据
```

```python
s=pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f'])
print(s)
#index自定义索引
```

```python
s=pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f'],name='a')
print(s)
#name：描述
```

```python
s=pd.Series({'a':1,'b':2,'c':3,'d':4,'e':5})
#可以通过字典来创建
print(s)
```

```python
s=pd.Series(s,index=['a','c'])
#在s中截取
print(s)
```

## 属性

```python
s=pd.Series([1,2,3],index=['a','b','c'],name='a')
print(s)
print(s.index) #索引
print(s.values) #元素
print(s.shape) #形状
print(s.ndim) #维度
print(s.size) #元素个数
print(s.dtype) #元素类型
print(s.name) #名字
```

```python
s=pd.Series([1,2,3],index=['a','b','c'],name='a')
print(s)
print(s.loc['a']) #显式索引
print(s.iloc[1]) #隐式索引
print(s.loc['a':'b']) #显式切片
print(s.iloc[0:2]) #隐式切片
print(s.at['a']) #显式索引，不支持切片
print(s.iat[0]) #隐式索引，不支持切片
```

## 访问数据

```python
s=pd.Series([1,2,3],index=['a','b','c'],name='a')
print(s['a'])  #直接访问
print(s[s<3])  #筛选
```

```python
s=pd.Series([1,2,3,4,5,6,7,8,9],index=['a','b','c','d','e','f','g','h','i',],name='a')
print(s)
print(s.head()) #返回前五行
print(s.tail()) #返回后五行
#head/tail的括号中写数字则返回前几行或后几行
```

## 常用方法

```python
s=pd.Series([1,2,3,4,5,6,7,8,9],index=['a','b','c','d','e','f','g','h','i',],name='a')
print(s)
print(s.isin([1,4,7]))   #判断几个值是否包含
print(s.isna())  #判断是不是缺失值
#sun()求和；mean()平均值；min()最小值；max()最大值
#var()方差；std()标准差；median()中位数；mode()众数；quantile()分位数
```

```python
s=pd.Series([1,2,3,4,5,6,7,8,9],index=['a','b','c','d','e','f','g','h','i',],name='a')
print(s)
print(s.describe()) #返回常见统计数据
```

```python
s=pd.Series([1,2,3,4,5,6,7,8,9],index=['a','b','c','d','e','f','g','h','i',],name='a')
print(s)
print(s.value_counts()) #每个唯一值出现次数
print(s.count())     #非缺失值数量
```

```python
print(s)
print(s.nunique()) #唯一值个数
print(s.unique()) #返回去重后数组
print(s.keys()) #获取索引
```

```python
s=pd.Series([1,1,3,5,5,6,7,7,9],index=['a','b','c','d','e','f','g','h','i',],name='a')
print(s)
print(s.drop_duplicates()) #去除重复项
```

```python
print(s.sample())  #随机抽一个元素
print(s.sort_index()) #按索引顺序排序
print(s.sort_values()) #按值排序
```

# DateFrame

## 创建

```python
s1=pd.Series([1,2,3,4,5,6,6,6])
s2=pd.Series([7,8,9,10,11,11,12,12])
s=pd.DataFrame({'一':s1,'二':s2})
s
```

```python
s=pd.DataFrame(
    {
        'name':['a','b','c'],
        'age':[10,11,12],
        'score':['1.1','1.2','1.3'],
        'bool':[True,False,True]
    },index=[1,2,3],columns=['name','score','age','bool']
)
#字典创建
#index索引；columns指定列的顺序，
s
```

## 属性

```python
s=pd.DataFrame(
    {
        'name':['a','b','c'],
        'age':[10,11,12],
        'score':['1.1','1.2','1.3'],
        'bool':[True,False,True]
    },index=[1,2,3],columns=['name','score','age','bool']
)
print(s.index) #行索引
print(s.columns) #列标签
print(s.values) #值
print(s.ndim) #维度
print(s.shape) #形状
print(s.dtypes) #元素类型
print(s.size) #元素个数
```

```python
print(s.T) #转置
#转置会让行索引和列标签互换
```

```python
print(s.loc[1])  #显式索引，获取行
print(s.iloc[1]) #隐式索引，获取行
print(s.loc[:,'name']) #显式索引，获取列
print(s.iloc[:,0])  #隐式索引，获取列
```

```python
print(s.at[1,'name']) #显式索引，获取单个元素
print(s.iat[0,0]) #隐式索引，获取单个元素
#获取单个元素用loc也可以
```

## 访问数据

```python
s=pd.DataFrame(
    {
        'name':['a','b','c'],
        'age':[10,11,12],
        'score':['1.1','1.2','1.3'],
        'bool':[True,False,True]
    },index=[1,2,3],columns=['name','score','age','bool']
)
s
```

## 常用方法

```python
#head()；tail()
s=pd.DataFrame(
    {
        'name':['a','b','c'],
        'age':[10,11,12],
        'score':['1.1','1.2','1.3'],
        'bool':[True,False,True]
    },index=[1,2,3],columns=['name','score','age','bool']
)
s
```

```python
print(s.isin([11])) #查看某个元素在不在
print(s.isna())  #查有没有空缺
```

```python
print(s.sum())  #各列求和
print(s['age'].sum()) #指定列求和
```

```python
# print(s.age.max()) #大
# print(s.age.min()) #小
# print(s.age.mean()) #平均
# print(s.age.median())#中位
# print(s.age.mode()) #众数
# print(s.age.std()) #标准差
# print(s.age.var()) #方差
print(s.describe()) #普遍信息
```

```python
print(s.count())  #每列非缺失值个数
```

```python
print(s.value_counts()) #每条信息出现次数
print(s.drop_duplicates()) #去重
print(s.duplicated(subset='age')) #检查是否重复(subset指定某一列
```

```python
print(s.replace(12,30))  #替换某个值
```

```python
print(s.cumsum()) #累计和
```

```python
print(s.sort_index())  #按索引排序
print(s.sort_values(by=['age','score'],ascending=[True,False]))
#按值排序
#by=必须写，指定排序的列，写多个按从前往后排（前一个相同再看下一个）
#ascending=指定正逆(True从小到大，False从大到小），by写了几个这也要写几个
```

```python
print(s.nlargest(2,['age'])) #取age列里最大的两个
print(s.nsmallest(2,['age'])) #取age列里最小的两个
```

## 缺失值处理

```python
s=pd.Series([1.2,1,'a',np.nan,None,pd.NA])#np.nan不是数字；None空；pd.NA不存在（excel的空单元格）
print(s)
```

```python
#查看是否是缺失值
print(s.isna())
print(s.isnull()) #一样的
```
