## 缺失值处理

```python
s=pd.Series([1.2,1,'a',np.nan,None,pd.NA])#np.nan不是数字；None空；pd.NA不存在（excel的空单元格）
print(s)
```

```python
#查看是否是缺失值
print(s.isna())
print(s.isnull()) #一样的
```

```python
s=pd.DataFrame(
    {'a':[1,2,3,2,5],
     'b':[1,3,None,2,3],
     'c':[np.nan,2,pd.NA,5,2],
     'd':[2,np.nan,None,pd.NA,1]})
s
```

```python
#查
print(s.isna())
print(s.isnull().sum()) #查看缺失值个数
```

```python
#删
print(s.dropna()) #剔除含缺失值的行
print(s.dropna(how='all'))  #剔除所有值都缺失的行
print(s.dropna(thresh=2)) #剔除非缺失值小于两个的行
print(s.dropna(axis=1))  #写0剔除含缺失值的行，写1剔除含缺失值的列
print(s.dropna(subset=['a','b'])) #剔除a列b列中含空缺值的行
```

```python
#填
print(s.fillna({'d':99}))  #在d列的空缺值中填上99
print(s.fillna(s['c'].mean()))  #一层中括号：用c列的平均值来填充空缺值：填充表中所有空缺值
print(s.fillna(s[['c']].mean()))#两层中括号：用c列的平均值填充c列的空缺值，只填充c列的空缺值
```

```python
print(s.ffill()) #用上面的相邻值填充
print(s.bfill()) #用下面的相邻值填充
```

## 分组聚合

```python
#组。groupby('分组的字段')['聚合的字段'].聚合函数()
s=pd.DataFrame({
    'id':[1,1,1,2,3,4,4,5,5,6,6,7],
    'select':[100,300,200,500,400,200,300,400,450,500,200,700]
})
s
```

```python
s.groupby('id').groups   #查看分组
```

```python
s.groupby('id').get_group(1) #查看某个具体的分组
```

```python
s1=s.groupby('id')[['select']].mean()   #id列相同的分一组，每组查select列的平均数
s1=s1.reset_index()
s1.sort_values('select',inplace=True) #从小到大排序
s1
```

```python
s=pd.DataFrame({
    'id1':[1,1,1,2,3,4,4,5,5,6,6,7],
    'id2':[1,1,2,1,1,1,2,1,2,1,2,1],
    'select':[100,300,200,500,400,200,300,400,450,500,200,700],
})
s
```

```python
s1=s.groupby(['id1','id2'])[['select']].mean()
s1['select']=s1['select'].round(1)   #保留一位小数
s1.sort_values('select',inplace=True)
s1
```
