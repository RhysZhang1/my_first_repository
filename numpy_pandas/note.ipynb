```python
import numpy as np
#numpy数组：ndarry
#多维，同质（所有元素类型必须一致），高效
```

# 一.ndarray

```python
a=[[10,20,30,40],[50,60,70,80]]
a=np.array(a,dtype=int)
#转换为np的数组,dtype定义数据类型：int,float,等
print(a)
```

## 同质

```python
a=np.array([1,'hi'])
#不同类型的会被强制转换（字符串>数字；浮点数>整数）
print(a)
```

## 属性

### 常见

```python
print('维度：',a.ndim) #numpy of dimensions
print('数组大小：',a.shape)
print('元素个数：',a.size)
print('元素类型：',a.dtype)
#数组名.ndim   返回数组维度
#0维:标量；1维:向量；2维:矩阵；大于3维:张量
#数组名.shape  返回数组行列
#数组名.size   返回数组元素数量
#数组名.dtype  返回数组元素类型
```

### 不常见

```python
print('转置',a.T)
print('单个元素占用的内存：',a.itemsize)
print('数组总内存占用：',a.nbytes)
print('内存存储方式',a.flags)
```

## 生成

```python
b=np.copy(a)
#复制数组
#深复制，两者互不影响了
print(b)
```

```python
a=np.zeros((3,4),dtype=int)
#生成一个3行4列全为零的数组(zero全为0，ones全为1，empyt未初始化的)
#一维数组写作(x,)
print(a)
```

```python
a=np.full((3,4),5,dtype=float)
#用5来填满一个三行四列的数组
print(a)
```

```python
b=np.zeros_like(a,dtype=int)
#形状和a相同，元素全为0
b=np.ones_like(a,dtype=int)
b=np.empty_like(a,dtype=int)
b=np.full_like(a,2,dtype=int)
print(b)
```

```python
a=np.arange(0,10,1).reshape(2,5)
#用法等同于range,reshape规定行列
print(a)
```

```python
#等间隔数列
a=np.linspace(0,11,12,dtype=int).reshape(3,4)
#生成一个第一个数为0,最后一个数为11,有12个数的数列,reshape规定行列
print(a)
```

```python
#等比数列
a=np.logspace(0,11,12,base=2,dtype=int).reshape(3,4)
#第一个数2的0次方，最后一个数2的4次方，12个数
#不写base默认10
print(a)
```

### 特殊矩阵

```python
#单位矩阵
a=np.eye(4,3,dtype=int)
#前两个参数：行，列
print(a)
```

```python
#对角矩阵
a=np.diag([4,5,1,2,3,4])
#生成矩阵的大小根据给的元素个数来
print(a)
```

```python
#随机矩阵（均匀分布）
a=np.random.rand(3,4)
#rand:生成0到1之间的随机数，括号中规定行列
print(a)
a=np.random.uniform(1,6,(3,4))
#生成1到6之间的随机浮点数，括号规定行列
print(a)
a=np.random.randint(3,20,(2,3))
#生成3到20之间的随机整数，括号内规定行列
print(a)
```

```python
#随机矩阵（正态分布）
a=np.random.randn(2,3)
#按标准正态分布，括号规定行列
print(a)
```

```python
np.random.seed(20)#设置随机数种子
a=np.random.randint(1,10,(2,3))
#随机数种子相同，则生成的数相同
print(a)
```

## 数据类型

```python
#bool  布尔（True/False）
#int  (8,16,32,64)  有符号整形
#uint (8,16,32,64)  无符号整形
#数字表示位数（8位，16位，32位，64位）（1，2，4，8字节）
#float (16,32,64)  浮点数  （半精度，单精度，双精度）
#complex  (64,128) 复数
a=np.array([1,0,1,1,0],dtype=bool)
print(a)
```

## 索引与切片

```python
#基本(一维)
a=np.random.randint(1,10,(10,))
print(a)
#与py的列表类似
print(a[2])
print(a[:])
print(a[2:5])
#过滤
print(a[a>5])
print(a[(a>2)&(a<8)])  #&且   |或
#slice函数
print(a[slice(2,8,2)])  #格式类似range
```
