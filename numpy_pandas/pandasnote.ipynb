# Series

## 基本

```python
s=pd.Series([1,2,3,4,5,6])
print(s)
#第一列索引，第二列数据
```

```python
s=pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f'])
print(s)
#index自定义索引
```

```python
s=pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f'],name='a')
print(s)
#name：描述
```

```python
s=pd.Series({'a':1,'b':2,'c':3,'d':4,'e':5})
#可以通过字典来创建
print(s)
```

```python
s=pd.Series(s,index=['a','c'])
#在s中截取
print(s)
```

## 属性

```python
s=pd.Series([1,2,3],index=['a','b','c'],name='a')
print(s)
print(s.index) #索引
print(s.values) #元素
print(s.shape) #形状
print(s.ndim) #维度
print(s.size) #元素个数
print(s.dtype) #元素类型
print(s.name) #名字
```

```python
s=pd.Series([1,2,3],index=['a','b','c'],name='a')
print(s)
print(s.loc['a']) #显式索引
print(s.iloc[1]) #隐式索引
print(s.loc['a':'b']) #显式切片
print(s.iloc[0:2]) #隐式切片
print(s.at['a']) #显式索引，不支持切片
print(s.iat[0]) #隐式索引，不支持切片
```
